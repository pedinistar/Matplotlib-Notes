```python
import matplotlib.pyplot as plt
```

```python
# importing numpy and pandas for working with array and running the code faster
import numpy as np
import pandas as pd
```

```python
# BASIC GRAPH
```

```python
# PLOTTING SIMPLE GRRAPH
x = [5,10,15,20,25]             # x axis in list
y = [6,12,18,24,30]             # y axis in list

plt.plot(x,y)
# (plt.plot) is used for plotting axis in graph
```

```python
x = np.array([0, 6])
y = np.array([0, 150])

plt.plot(x,y)
plt.show()
```

```python
# for changing color of the line in graph
plt.plot(x,y,color = 'yellow')
```

```python
# for marking the graph
x = [5,10,15,20,25]
y = [6,12,18,24,30]

# for marking [marker = ''] function is used
plt.plot(x,y,color= 'yellow',marker = '*')
```

```python
# coloring edge of the marker on graph
# use [markeredgecolor = ''] function for coloring edge of the marker
plt.plot(x,y,color= 'yellow',marker = '*',markeredgecolor = 'blue')
```

```python
# increasing edge size of marker
# use [markersize = num] function for increasing edge size of marker
plt.plot(x,y,color= 'yellow',marker = '*',markeredgecolor = 'blue',markersize = 10)
```

```python
# changing style of line in graph
# for changing style of line use [linestyle = '']
plt.plot(x,y,color= 'm',marker = '*',linestyle = ':',markeredgecolor = 'blue',markersize = 10)
# supported values for linestyle are ['-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted']
```

```python
# short form (color, marker, linestyle)
plt.plot(x,y,'r*--')
```

```python
# heading the graph
plt.plot(x,y,'r*--')
plt.title('Heading1')
# [title] method is used for titling/heading the graph
```

```python
# change fontsize of the heading
plt.plot(x,y,'r*--')
plt.title('Heading1',fontdict={'fontsize': 20})
```

```python
# labeling x and y axis
plt.plot(x,y,'r*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('First Graph',fontdict={'fontsize': 20})
# use show() function for showing changes in graph
plt.show()
```

```python
# legend of the graph
plt.plot(x,y,'b*--')
# legend is used to identify elements of the plot in graph
plt.legend(['single element'])
plt.show()
```

```python
# save the graph
plt.plot(x,y,'b*--')
plt.savefig('First Graph',dpi = 200)
```

```python
# LINE GRAPH
```

```python
# creating data for line graph
dict = {'year': [2000,2001,2002,2003,2004,2005],
        'india': [5000,6000,7000,8000,9000,10000],
        'china': [4000,5000,6000,7000,8000,9000],
        'america': [8000,9000,15000,18000,20000,20000],
        'japan': [3000,4000,8000,9000,15000,20000]
       }
```

```python
mobile = pd.DataFrame(dict)
mobile
```

```python
# creating csv file for data
mobile.to_csv('mobile.csv')
```

```python
# importing csv file
mobile = pd.read_csv('mobile.csv')
mobile.head()
```

```python
# PLOTTING DATA INTO THE GRAPH
plt.plot(mobile.year,mobile.india,'b.-',label = 'india')
plt.plot(mobile.year,mobile.china,'r.-',label = 'china')
plt.plot(mobile.year,mobile.america,'g.-',label = 'america')
plt.plot(mobile.year,mobile.japan,'m.-',label = 'japan')
plt.xlabel('year')
plt.ylabel('price')
plt.title('Mobile Price Year Wise',fontdict={'fontsize':20})
plt.legend()
plt.savefig('Mobile Price.png',dpi = 300)
plt.show()
```

```python
# HISTOGRAM
```

```python
# PLOTTING HISTOGRAM
bunch = [0,20,40,60,80,100]
x = [10,20,30,40,50,60]

plt.hist(x,bins=bunch,color = 'black')
plt.xlabel('Bunch')
plt.ylabel('Data')
plt.legend(['ok'])
plt.show()
```

```python
# PIECHART
```

```python
# PLOTTING PIE CHART
y = np.array([20,30,35,40])
name = ['Ajay','Udey','Suresh','Sanjay']

plt.pie(y,labels=name)
plt.show()
```

```python
# EXPLODE THE HIGHER AMOUNT OF DATA
# Explode is used for separating slice of a piechart
nameexplode = [0,0,0,0.2]
plt.pie(y,labels=name,explode = nameexplode)
plt.show()
```

```python
# creating shadow of piechart
plt.pie(y,labels=name,explode = nameexplode,shadow = True)
plt.savefig('PieChart.png',dpi = 300)
plt.show()
```

```python
# coloring piechart
mycolor = ['b','hotpink','black','m']
plt.pie(y,labels=name,explode = nameexplode,shadow = True,colors = mycolor)
plt.show()
```

```python
# legend of piechart
mycolor = ['b','hotpink','black','m']
plt.pie(y,labels=name,explode = nameexplode,shadow = True,colors = mycolor)
plt.legend()
plt.savefig('PieChart2.png',dpi=300)
plt.show()
```

```python
# PLOTTING BOXPLOT
test = np.random.normal(200,50,400)
plt.boxplot(test)
plt.show()
```

```python
# PLOTTING MULTIPLE BOXPLOT
test1 = np.random.normal(200,50,400)
test2 = np.random.normal(100,70,300)
test3 = np.random.normal(150,100,200)
test4 = np.random.normal(250,50,400)

data = [test1,test2,test3,test4]
plt.boxplot(data)
plt.show()
```
